```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from tueplots import bundles
```

**Delhi-NCR active learning**

```python
plt.rcParams.update(bundles.icml2022())
# Read data from CSV files
data_files1 = ['India_bald/f1_score_bald.csv', 'India_bald/precision_bald.csv', 'India_bald/recall_bald.csv', 'India_bald/test_accuracy_bald.csv']  # Replace with your file names
data1 = [np.array(pd.read_csv(file)) for file in data_files1]
data_files2 = ['India_entropy/f1_score_entropy.csv', 'India_entropy/precision_entropy.csv', 'India_entropy/recall_entropy.csv', 'India_entropy/test_accuracy_entropy.csv']  # Replace with your file names
data2 = [np.array(pd.read_csv(file)) for file in data_files2]
random_data = pd.read_csv('India_random/India_random_compiled.csv')
data3 = random_data.reset_index(drop=True)

# Create a 2x2 subplot grid
fig, axs = plt.subplots(2, 2, figsize=(5,4))
titles = ['F1', 'precision', 'recall', 'accuracy']
# Plot data on each subplot
for i, ax in enumerate(axs.flatten()):
    random_mean = np.array(data3.iloc[2*i])[1:]
    random_sigma = np.array(data3.iloc[2*i+1])[1:]
    random_mean = random_mean.astype(float)
    random_sigma = random_sigma.astype(float)
    ax.plot(data1[i], label='BALD', linewidth = 1)
    ax.plot(data2[i], label='Entropy', linewidth = 1)
    # ax.plot(data4[i], label = 'Deterministic entropy', color = 'red', linewidth =1)
    ax.plot(random_mean, color = 'green', label = r'Random $(\mu\pm\sigma)$', linewidth = 1)
    ax.fill_between(np.arange(len(random_mean)), random_mean-random_sigma, random_mean+random_sigma, color = 'green', alpha = 0.2)
    # ax.set_title(titles[i])
    ax.set_ylabel(titles[i])
    if(i==2 or i==3):
        ax.set_xlabel('Iterations')
    if(i==0):
        ax.legend(frameon = False, )
    sns.despine(ax=ax) 

# Adjust layout and display the plots
plt.savefig('figures/India_AL.pdf')
plt.show()
```
